```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n1, h1 = np.loadtxt('depend/bw_W', unpack=True)
n2, h2 = np.loadtxt('depend/grey_W', unpack=True)
n3, h3 = np.loadtxt('depend/rgb_W', unpack=True)

plt.figure(figsize=(10, 6))

plt.plot(n1, h1, 'b-o', linewidth=1, markersize=4, label='Для ЧБ изображений')
plt.plot(n2, h2, 'r-o', linewidth=1, markersize=4, label='Для изображений в оттенках серого')
plt.plot(n3, h3, 'g-o', linewidth=1, markersize=4, label='Для цветных изображений')


plt.xlabel('Максимальный размер словаря (байт)')
plt.ylabel('Коэффициент сжатия')
plt.title('Зависимость коэффициента сжатия от размера словаря')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
